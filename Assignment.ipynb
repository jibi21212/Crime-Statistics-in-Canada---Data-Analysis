## Question:
What factors contribute to the rise and fall of crime rates in Canada, and what implications do these trends have for our understanding of crime dynamics in the country?

```python
import pandas as pd
import matplotlib.pyplot as plt
crime_data = pd.read_csv('https://raw.githubusercontent.com/Crime-Dataset/Crime-Analysis/main/Reported_Crimes.csv')
print("Original Dataset:")
print(crime_data.head())
# Remove the 'ObjectId' column
crime_data_cleaned = crime_data.drop(columns=['ObjectId','Index_'])


```

## Cleaned Dataset

```python
# Display the cleaned dataset
print("\nCleaned Dataset:")
print(crime_data_cleaned.head())
```

```python
# Plotting crime rates over time as a bar graph
plt.figure(figsize=(12, 6))
plt.bar(crime_data_cleaned['ReportedYear'], crime_data_cleaned['Count_'], color='red')
plt.title('Crime Rates in Canada Over Time')
plt.xlabel('Reported Year')
plt.ylabel('Crime Rate')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

The graph illustrates noticeable variations in crime rates from year to year.
Certain years exhibit higher bars, indicating elevated crime rates, while others showcase lower bars, suggesting periods of reduced reported criminal activities.
Peaks in the graph represent years with higher crime rates, suggesting potential spikes in criminal activities during those periods.
Valleys, on the other hand, denote years with lower crime rates, indicating potential periods of decreased reported criminal incidents.


```python
# Grouping data by ReportedYear to analyze counts of reported and cleared crimes
crime_counts_by_year = crime_data_cleaned.groupby('ReportedYear')[['Count_', 'CountCleared']].sum()

# Plotting the counts of reported and cleared crimes over the reported years
plt.figure(figsize=(12, 6))
plt.plot(crime_counts_by_year.index, crime_counts_by_year['Count_'], marker='o', label='Reported Crimes', linestyle='-', color='blue')
plt.plot(crime_counts_by_year.index, crime_counts_by_year['CountCleared'], marker='o', label='Cleared Crimes', linestyle='-', color='green')
plt.title('Comparison of Reported and Cleared Crimes Over Time')
plt.xlabel('Reported Year')
plt.ylabel('Crime Count')
plt.legend()
plt.grid(True)
plt.show()
```

Temporal Patterns: The graph's two lines distinctly exhibit temporal patterns in reported and resolved crimes throughout the documented years. The line depicting reported crimes showcases the varying frequency of criminal incidents, providing valuable perspectives on the overall crime scenario.

Discrepancies and Associations: Differences between the two lines highlight intervals where reported crimes exceed cleared crimes or vice versa. These variations may indicate obstacles or triumphs in law enforcement endeavors. Associations between the lines of reported and cleared crimes may reflect the responsiveness of law enforcement to reported incidents.

It doesn't seem absurd that the number of reported crimes is always greater than the number of cleared crimes. However, there seems to be a massive disparity which suggests there is room for improvement with the law enforcement agency.

```python
# Grouping data by GeoDivision to analyze counts of cleared crimes
cleared_crimes_by_location = crime_data_cleaned.groupby('GeoDivision')['CountCleared'].sum().sort_values(ascending=False)

# Plotting the top GeoDivisions with the highest counts of cleared crimes
plt.figure(figsize=(12, 6))
cleared_crimes_by_location.head(10).plot(kind='bar', color='green')
plt.title('Top 10 GeoDivisions with Highest Counts of Cleared Crimes')
plt.xlabel('GeoDivision')
plt.ylabel('Cleared Crimes Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

The bar graph showcasing the top 10 GeoDivisions with the highest counts of cleared crimes provides a snapshot of law enforcement efficacy across different regions:
- The graph highlights the GeoDivisions with the most successful resolutions or closures of reported crimes, as indicated by the counts of cleared crimes.
- A closer examination of these high-performing GeoDivisions could reveal factors contributing to their success. These factors may include robust law enforcement strategies, community engagement, or specialized crime-solving units.
